```python
medidas_tendencia = {
    'Media': df1[columnas_numericas].mean(),
    'Mediana': df1[columnas_numericas].median(),
    'Moda': df1[columnas_numericas].mode().iloc[0]  # Toma la primera moda si hay más de una
}

# Mostrar resultados en consola
print("Medidas de tendencia central:")
print("Media:\n", medidas_tendencia['Media'])
print("\nMediana:\n", medidas_tendencia['Mediana'])
print("\nModa:\n", medidas_tendencia['Moda'])
```

```python

# Columnas numéricas
columnas_numericas = ['Age', 'Balance..euros.', 'Last.Contact.Day', 
                      'Last.Contact.Duration', 'Campaign', 'Pdays', 'Previous']

# Generar un diagrama de caja para cada columna numérica
for col in columnas_numericas:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df1[col])
    plt.title(f"Diagrama de Caja de {col}")
    plt.ylabel(col)
    plt.show()
```

```python
# Definir las columnas categóricas de interés
columnas_categoricas = ['Job', 'Marital.Status', 'Education', 'Credit', 
                        'Housing.Loan', 'Personal.Loan', 'Contact', 
                        'Last.Contact.Month', 'Poutcome']

# Convertir 0 y 1 en la columna 'Subscription' a 'No' y 'Sí'
df1['Subscription'] = df1['Subscription'].replace({0: 'No', 1: 'Sí'})

# Construcción de tablas de contingencia
tablas_contingencia = {}
for col in columnas_categoricas:
    # Tabla de conteo
    tabla_conteo = pd.crosstab(df1[col], df1['Subscription'])
    # Tabla de porcentaje (por fila)
    tabla_porcentaje = pd.crosstab(df1[col], df1['Subscription'], normalize='index') * 100
    
    # Guardar en un diccionario
    tablas_contingencia[col] = {
        'Conteo': tabla_conteo,
        'Porcentaje': tabla_porcentaje
    }
    
    # Imprimir las tablas de contingencia
    print(f"Tablas de contingencia para '{col}':\n")
    print("Conteo:\n", tabla_conteo)
    print("\nPorcentaje:\n", tabla_porcentaje)
    print("\n" + "="*50 + "\n")
```

```python



# Crear una matriz de gráficos de dispersión
sns.pairplot(df1[columnas_numericas])
plt.suptitle('Matriz de Gráficos de Dispersión', y=1.02)  # Ajusta el título
plt.show()
```

```python
# Calcular la matriz de correlación
matriz_correlacion = df1[columnas_numericas].corr()

# Mostrar la matriz de correlación
print("Matriz de Correlación:\n", matriz_correlacion)

# visualizar la matriz de correlación con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Matriz de Correlación')
plt.show()
```
